```python
#用pandas来操作数据
import pandas as pd
#核心算法PCA
from sklearn.decomposition import PCA
#可视化
#import matplotlib.pyplot as plt
#import seaborn as sns
#替换数据源
#pandas可以读取和写入Oracle、MySql、MangoDB
data = pd.read_csv('heart.csv')#数据源换成你用的数据表
data.head()
```

```python
#创建PCA实例
pca = PCA(n_components = 1)
#拟合替换的数据源
pca.fit(data)
#查看权重
print(pca.components_)
```

```python
#如果样本特征量纲差异过大
#可以考虑使用MinMaxScaler先标准化
from sklearn.preprocessing import MinMaxScaler
data_scl = MinMaxScaler().fit_transform(data)
data_scl = pd.DataFrame(data_scl)
#data_scl.head()
```

```python
#再使用PCA拟合
pca2 = PCA(n_components = 1)
pca2.fit(data_scl)
pca2.components_#获得权重
```

```python
#计算总积分
score = pca2.transform(data_scl)
#如果积分有负数，也可以用MinMaxScaler处理到0-1
score_scl = MinMaxScaler().fit_transform(score)
#可以转为DataFrame，也可以是数组
#看你怎么方便操作
score_scl = pd.DataFrame(score_scl)
#下面这个是为了检查，可以注释掉
#score_scl.head()
```
