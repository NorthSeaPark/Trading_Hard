```python
#导入jqdata和datetime
from jqdata import *
import datetime
```

```python
#使用timedelta计算出前一日的日期
yesterday = datetime.date.today() - datetime.timedelta(days=2)
#创建query，获取前一日的数据
q = query(finance.CCTV_NEWS).filter(finance.CCTV_NEWS.day==yesterday)
#执行query
news = finance.run_query(q)
#检查数据的前5行
news.head(5)
```

```python
news.head(20)
```

```python
#通过指定行与列序号，
#可以查看某一条新闻的完整正文
text= news.iloc[13,3]
#显示文本
text
```

```python
#去掉正文中的空格
text = text.replace(' ','')
#去掉回撤与换行\r\n
text = text.replace('\r\n','')
#检查结果
text
```

```python
#导入结巴分词
import jieba
#使用cut方法即可完成分词
words = jieba.cut(text)
#在词之间插入空格
words = ' '.join(words)
#检查结果
words
```

```python
type(words)
```

```python
#使用lcut将文本分词并存为列表
word_list = jieba.lcut(text)
#检查列表中前30个元素
word_list[:30]
```

```python
#首先创建一个空字符串
word = ''
#逐行读取停用词表，并按行分隔后，存入列表中
stopwords = [line.strip() for line in open('stopwords.txt',encoding='UTF-8').readlines()]
#对于原文本中的元素
for element in words:
    #如果不在停用词表中
    if element not in stopwords:
        #就添加到我们创建的空字符串中
        word += element
#检查一下结果
word
```

```python
#导入结巴分词中的analyse
import jieba.analyse
#继续使用停用词表
jieba.analyse.set_stop_words("stopwords.txt")
```

```python
#抽取关键词
#withWeight参数控制的是，返回的结果是否包含关键词的权重
tags = jieba.analyse.extract_tags(text,withWeight=False)
#检查结果
tags
```

```python
#调节extract_tags的参数
#设置topK，也就是关键词数量为10
#withWeight为True，让结果带上权重数值
tags = jieba.analyse.extract_tags(text,
                                  topK = 10,
                                  withWeight=True)
#检查结果
tags
```
